```python
import pandas as pd
import os
import matplotlib.pyplot as plt

os.chdir(r"C:\Users\Chris\Documents\DS Classes and Scripts\DSC530_EDA")
nio = pd.read_csv("NIO.csv")
nio
```

```python
nio = nio[:600]
nio
```

```python
tesla = pd.read_csv("TSLA.csv")
tesla
```

```python
tesla = tesla[:600]
tesla
```

```python
df = pd.concat([nio, tesla],axis=1)
df
```

```python
df.columns = ['Index', 'Nio Open', 'Nio High', 'Nio Low', 'Nio Close', 'NioAdjCl', 'Nio Vol', 'Date', 'Tesla Open',
             'Tesla High', 'Tesla Low', 'Tesla Close', 'TAdjCl', 'Tesla Vol']
df
```

```python
# Import library and dataset
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
#df = sns.load_dataset('')
 
# Method 1: on the same Axis
sns.distplot(df["Nio High"], color="skyblue", label="Nio High", fit=norm)
sns.distplot(df["Tesla High"], color="red", label="Tesla High", fit=norm)
plt.legend()
plt.xlabel("Daily High Price")
plt.ylabel("Count")

```

```python
# library and data
import seaborn as sns
import matplotlib.pyplot as plt
#df = sns.load_dataset('iris')
 
# plot
sns.distplot( df["Nio Low"] , color="skyblue", label="Nio Low", fit=norm)
sns.distplot( df["Tesla Low"] , color="red", label="Tesla Low", fit=norm)
plt.legend()
plt.xlabel("Daily Low Price")
plt.ylabel("Count")
```

```python
# plot

sns.distplot( df["Nio Open"] , color="skyblue", label="Nio Open", fit=norm)
sns.distplot( df["Tesla Open"] , color="red", label="Tesla Open", fit=norm)
plt.legend()
plt.xlabel("Daily Opening Price")
plt.ylabel("Count")
```

```python
sns.distplot( df["Nio Close"] , color="skyblue", label="Nio Close", fit=norm)
sns.distplot( df["Tesla Close"] , color="red", label="Tesla Close", fit=norm)
plt.legend()
plt.xlabel("Daily Closing Price")
plt.ylabel("Count")
```

```python
sns.distplot( df["Nio Vol"] , color="skyblue", label="Nio Vol", fit=norm)
sns.distplot( df["Tesla Vol"] , color="red", label="Tesla Vol", fit=norm)
plt.legend()
plt.xlabel("Daily Volume of Shares Traded")
plt.ylabel("Count")
```
