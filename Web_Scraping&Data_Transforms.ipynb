```python

from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
import lxml.html as lh
import csv
import requests
import urllib.request
import matplotlib.pyplot as plt
```

```python
url = "https://carsalesbase.com/china-nio/"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())
```

```python
#Looking for all tables on the website
all_tables=soup.find_all("table")
all_tables
table = soup.find('table',{'id':'table_1'})
table
```

```python
data = [ [ td.text for td in tr.find_all("td") ] for tr in table.find_all("tr")]
data
```

```python
# Saved the scraped table data into a pandas dataframe

df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
# Renamed the headers

df.columns = ['Year', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df
```

```python
# Ditched the 1st row below the headers since it was filled with 'None'
df = df.dropna()
df
```

```python
# Threw out the last row as well displaying average monthly sales, will replace with something better
df = df.drop([4])
df

```

```python
type(df)
```

```python
#blanks were throwing the errors for conversion so replace as NaN
df = df.replace(r'^\s*$', np.nan, regex=True)
df
```

```python
#convert to float -  Thank you for the help with this portion. I know how powerful regex is and I need to focus on those
df.loc[df.any(axis=1)] = df.replace('[\$,]', '', regex=True).astype(float).fillna(0)
df

```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df
```

```python
print('Dataframe: ')
print(df)
print()
print('Dtypes: ')
print(df.dtypes)
print()
```

```python
# Added the sum of all the months together for a quick look at what the most productive months are. 

df.loc['Total']= df.sum()
df
```
