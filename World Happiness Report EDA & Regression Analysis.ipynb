```python
from sklearn.linear_model import LinearRegression

X = whr_df[["GDP per capita"]]
X.head
```

```python
y = whr_df[["Score"]]
y.head
```

```python
reg = LinearRegression()
model = reg.fit(X,y)
print("intercept: ", model.intercept_)
print("coef: ", model.coef_)
print("rscore. ", model.score(X,y))
```

```python
def linear_reg(col,text,prdctn):
    
    X = whr_df[[col]]
    y = whr_df[["Score"]]
    reg = LinearRegression()
    model = reg.fit(X,y)
    
    
    plt.figure(figsize=(12,6))
    g = sns.regplot(x=whr_df[col],y=whr_df["Score"],ci=None,scatter_kws = {'color':'r','s':9})
    g.set_title("Prediction Model")
    g.set_ylabel("Score")
    g.set_xlabel(col)
    plt.show()
    
    print(text,": ", model.predict([[prdctn]]))
```

```python
linear_reg("Social support","The happiness value of a country whose social support value is 1.5",1.5)
```

```python
linear_reg("GDP per capita","The happiness value of a country whose GDP per capita value is 1.5",1.5)
```

```python
linear_reg("Generosity","The happiness value of a country whose Generosity value is 0.8",0.8)
```

```python
linear_reg("Perceptions of corruption","The happiness value of a country whose Perceptions of corruption value is 1",1)
```

```python
linear_reg("Freedom to make life choices","The happiness value of a country whose Freedom to make life choices value is 0.75", 0.75)
```

```python
linear_reg("Healthy life expectancy","The happiness value of a country whose Healthy life expectancy value is 0.9",0.9)
```
